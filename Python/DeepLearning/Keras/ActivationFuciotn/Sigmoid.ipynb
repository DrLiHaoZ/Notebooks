```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf 
from tensorflow import keras
```

```python
X = np.arange(start= -25,stop = 25,step=0.1)
y1 = 1/(1 + np.exp(-X))
y2 = 1/(1 + np.exp(-X / 0.5))
plt.plot(X,y1,'ro',alpha = 0.2)
plt.plot(X,y2,'bo',alpha = 0.2)
```

```python
X = np.arange(start= -25,stop = 25,step=0.1)
y1 = 1/(1 + np.exp(-X))
y2 = tf.keras.activations.sigmoid(X).numpy()
plt.plot(X,y1,'ro',alpha = 0.2)
plt.plot(X,y2,'bo',alpha = 0.2)
```
