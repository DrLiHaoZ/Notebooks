```python
import matplotlib.pyplot as plt
import numpy as np
import  matplotlib.lines as lines
```

```python
a = np.array([[1,2],[2,1]])
fig,axs = plt.subplots()
axs.add_artist(lines.Line2D(xdata=[1,1,2],ydata=[1,2,3]))

axs.set_xlim(0,5)
axs.set_ylim(0,5)

plt.show()
```

```python
x = np.linspace(start=0,stop = 10,num = 100)
y = np.sin(x)

fig,axes = plt.subplots()
axes.add_artist(lines.Line2D(xdata=x,ydata=y))

axes.set_xlim(0,10)
axes.set_ylim(-1.5,1.5)

plt.show()
```

```python
x = np.linspace(start=0,stop = 10,num = 100)
y = np.sin(x)

fig,axes = plt.subplots()
axes.scatter(x,y)

axes.set_xlim(0,10)
axes.set_ylim(-1.5,1.5)
plt.show()
```
