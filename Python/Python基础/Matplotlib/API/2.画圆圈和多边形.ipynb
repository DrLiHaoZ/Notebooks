```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches
```

```python
fig,axes = plt.subplots(nrows=1,ncols = 1,figsize = (5,5))
circle = patches.Circle(xy=(0,0),radius=1,alpha = 0.5,color = "#aaaaaa")
axes.add_artist(circle)
axes.set_xlim(-2,2)
axes.set_ylim(-2,2)
plt.show()
```

```python
fig,axes = plt.subplots(nrows=1,ncols = 1,figsize = (5,5))
polygon = patches.Polygon(xy=np.array([[1,2],[0,0],[1,1]]),closed=True)
axes.add_artist(polygon)
axes.set_xlim(0,4)
axes.set_ylim(0,4)
plt.show()
```
