模拟卡方分布

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = np.random.normal(size=10000).reshape(1000,10)

chi2_1 = df[:,0] ** 2
chi2_2 = df[:,0] ** 2 + df[:,1] ** 2
chi2_3 = df[:,0] ** 2 + df[:,1] ** 2 + df[:,2] ** 2
```

```python
sns.kdeplot(chi2_1)
sns.kdeplot(chi2_2)
sns.kdeplot(chi2_3)
```
