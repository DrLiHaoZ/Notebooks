```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 取3个数均值的样本100个
a1 = np.random.randn(100,3)
a1_mean = np.mean(a1,axis=1)
#取5个数均值的样本100个
a2 = np.random.randn(100,5)
a2_mean = np.mean(a2,axis=1)

a3 = np.random.randn(100,7)
a3_mean = np.mean(a3,axis=1)

a4 = np.random.randn(100,14)
a4_mean = np.mean(a4,axis=1)

sns.kdeplot(a1_mean)
sns.kdeplot(a2_mean)
sns.kdeplot(a3_mean)
sns.kdeplot(a4_mean,color="black")
```

样本的统计规模约大，样本统计量的分布就越窄
