即使原始总体不符合正态分布，只要样本的规模足够大，那么多个样本的均值就会呈现出正太分布

```python
import numpy as np
import seaborn as sns
```

```python
a = np.random.exponential(scale=1,size = 1000)
sns.kdeplot(a)
```

```python
a = []
for i in range(10):
    a.append(np.mean(np.random.exponential(scale=1,size = 1000*i*5).reshape(1000,i*5),axis=1))
    sns.kdeplot(a[i])
```
