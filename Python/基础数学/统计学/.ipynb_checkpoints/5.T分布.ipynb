```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

$t = \frac{\bar{X}-\mu}{\sigma/\sqrt{n}}$

```python
#首先来模拟出自由度为5-1的t分布
np.random.seed(123)
df = np.random.randn(10000,5)
t5 = np.mean(df,axis=1) / (np.std(df,axis=1) / np.sqrt(5))
t5_gen = np.random.standard_t(df = 4,size = 10000)
sns.kdeplot(t5)
sns.kdeplot(t5_gen)
```

```python
#首先来模拟出自由度为10-1的t分布
np.random.seed(123)
df = np.random.randn(10000,10)
t10 = np.mean(df,axis=1) / (np.std(df,axis=1) / np.sqrt(10))
t10_gen = np.random.standard_t(df = 9,size = 10000)
sns.kdeplot(t10)
sns.kdeplot(t10_gen)
```
