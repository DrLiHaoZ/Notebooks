t versus $\chi^2$

if X1,...,Xn iid $N(\mu,\sigma^2)$,then:

$\frac{\bar{X}-\mu}{\sigma/\sqrt n}\sim N(0,1)$

但是$\sigma$未知的时候
则用样本的方差$\hat \sigma$来代替$\sigma$，则有：

$\frac{\bar X-\mu}{\hat \sigma / \sqrt n}\sim t_{n-1}$

而 $\hat \sigma = \sqrt{\frac{\sum(X_i - \bar X)^2}{n-1}}$  

$\frac{\bar X-\mu}{\hat \sigma / \sqrt n} = \frac{\bar X-\mu}{ \sigma / \sqrt n} \cdot \frac{1}{\frac{\hat \sigma}{\sigma}} $ = 

$Z \cdot \frac{1}{\sqrt{\frac{\sum(X_i - \bar X)^2}{(n-1)\sigma^2}}}$ = 

$\frac{Z}{\sqrt{\frac{\chi_{n-1}^{2}}{n-1}}}$

故有 $t^2=\frac{\chi^2_1}{\chi^2_n / n} = F(1,n)$

```python
df = np.random.f(1,10,size=10000)
df2_chi_nu = np.random.chisquare(df = 1,size=10000)
df2_chi_de = np.random.chisquare(df = 10,size=10000)
df2 = np.divide(df2_chi_nu,df2_chi_de/10)
sns.kdeplot(df)
sns.kdeplot(df2)
```
