```python
np.random.seed(123)
df = np.random.randn(100000,30)
t30 = np.mean(df,axis=1) / (np.std(df,axis=1) / np.sqrt(30-1))
```

```python
sns.kdeplot(t30)
```

```python
one_side = 1 - np.searchsorted(sorted(t30),2.7386) / 100000
print(one_side)
p_value = 2 * one_side
print(p_value)
```
