```python
import numpy as np
import random
import seaborn as sns
```

```python
a = np.random.normal(loc = 10,scale = 2,size = 10000)
b = np.random.normal(loc = 10,scale = 2,size = 10000)
sns.kdeplot(a)
sns.kdeplot(b)
```

```python
chi2_sim = ((a - b)**2 / a)
chi2 = (np.random.normal(loc = 0,scale=1,size = 1000))**2

sns.kdeplot(chi2_sim)
sns.kdeplot(chi2)
```

```python
c = a - b
sns.kdeplot(c)
print(np.std(c))
print(np.sqrt(4+4))
```

```python
c = (a - b )/np.sqrt(np.abs(a)) 
sns.kdeplot(c)
print(np.std(c))
```
