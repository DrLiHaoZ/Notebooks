```python
import numpy as np
import tensorflow as tf 
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = plt.imread("./IMG/IMG_Genshin_Impact.jpeg")
plt.imshow(img)
```

```python
red_channel, green_channel, blue_channel = img[:,:,0],img[:,:,1],img[:,:,2]
fig,axs = plt.subplots(nrows=3,ncols=1, figsize= (5,5))
axs[0].imshow(red_channel,cmap = "gray")
axs[1].imshow(green_channel, cmap="gray")
axs[2].imshow(blue_channel, cmap="gray")
red_channel.shape
```

```python
blue_channel_for_conv = tf.constant(blue_channel)
blue_channel_for_conv = tf.reshape(blue_channel_for_conv,shape = (1,1080,1920,1))
input_shape = blue_channel_for_conv.shape

conv = keras.layers.Conv2D(
        filters = 1,
        kernel_size = (3,3),
        strides = 1,
        activation ='relu', 
        padding ="valid", 
        input_shape = input_shape[1:]
)(blue_channel_for_conv)
```
