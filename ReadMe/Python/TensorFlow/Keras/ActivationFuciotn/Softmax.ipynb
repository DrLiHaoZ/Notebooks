- Softmax function convert a vector of values to a probability distribution

$ S_i = \frac{e^i}{\sum{_j}e^j}$

```python
import numpy as np 
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
a = np.random.binomial(n = 20, p  = 0.2,size= 1000)
plt.hist(a)
```

```python
softmaxed_a = np.exp(a) / np.sum(np.exp(a))
np.sum(softmaxed_a)
plt.hist(softmaxed_a,bins=10)
```

```python
inputs = tf.random.normal(shape=(32, 10))
outputs = tf.keras.activations.softmax(inputs)
print(outputs)
print(tf.reduce_sum(outputs,axis=1))
```

```python
a = tf.reshape(a,shape=(1,-1))
a = tf.cast(a,dtype=tf.float32)
output = keras.activations.softmax(a)
print(output)
print(tf.reduce_sum(output,axis = -1))
```
