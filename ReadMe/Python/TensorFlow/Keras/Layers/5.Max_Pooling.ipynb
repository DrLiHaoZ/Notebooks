```python
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
img = plt.imread("./IMG/IMG_Genshin_Impact.jpeg")
plt.imshow(img)
```

```python
red_channel, green_channel, blue_channel = img[:,:,0],img[:,:,1],img[:,:,2]
fig,axs = plt.subplots(nrows=3,ncols=1, figsize= (5,5))
axs[0].imshow(red_channel,cmap = "gray")
axs[1].imshow(green_channel, cmap="gray")
axs[2].imshow(blue_channel, cmap="gray")
red_channel.shape
```

- the original shape of the image is (1080, 1920), let's max-pooling it 

```python
blue_channel_for_pooling = tf.constant(blue_channel)
blue_channel_for_pooling = tf.reshape(blue_channel_for_pooling,shape = (1,1080,1920,1))
pooling = keras.layers.MaxPooling2D(pool_size = (5,5),strides  = (1,1),padding = "valid")

pooled = pooling(blue_channel_for_pooling)
pooled = tf.squeeze(input=pooled,axis = [0,-1])

fig,axs = plt.subplots(nrows = 2,ncols = 1,figsize = (15,15))
axs[0].imshow(blue_channel,cmap = "gray")
axs[1].imshow(pooled.numpy(),cmap = "gray")
```
