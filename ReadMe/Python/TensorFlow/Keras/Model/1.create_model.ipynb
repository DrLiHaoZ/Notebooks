```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# prepare the data of multivariable linear regression

np.random.seed(10)
DATA_ROW_NUM = 1000
DATA_COL_NUM = 2

x = np.random.rand(DATA_ROW_NUM,DATA_COL_NUM) * 100
print('the row #1 of x is {x_1}'.format(x_1 = x[0]))

# fake the W attribute
W = np.random.rand(DATA_COL_NUM)

y = x @ W
print(' x is {x}, \n w is {W}, \n y is {y}'.format(x = x[0],W = W,y = y[0]))

```

### two ways to create a model directly not from models.sequential
1. subclass the Model
2. use model directly


```python
class MyModel(tf.keras.Model):
    def __init__(self):
        super().__init__()
        self.input_layer = tf.keras.layers.Input(shape=(DATA_COL_NUM,))
```

```python
inputs = tf.keras.layers.Input(shape=(DATA_COL_NUM,))
hidden = tf.keras.layers.Dense(6,activation = tf.nn.relu)(inputs)
outputs = tf.keras.layers.Dense(1,activation = tf.nn.relu)(hidden)
model = tf.keras.Model(inputs = inputs, outputs = outputs)

model.compile(optimizer = "Adam", loss = "mse" , metrics = ['mae','mse'])
model.fit(x,y,epochs=100)
```

```python
print(x.shape)
print(y.shape)
model.summary()
```

```python
test = [[99,89]]

print(model.predict(test))
print(test[0] @ W)
```
