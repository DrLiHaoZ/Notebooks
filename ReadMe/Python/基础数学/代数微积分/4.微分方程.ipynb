```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

$$\frac{dy}{dx} = -0.03y $$
$$\int \frac{1}{y} dy = \int dt (-0.03)$$
$$lny = -0.03 t + C$$

```python
def model(y,t):
    k = 0.03
    dydt = - k * y 
    return dydt

y0 = 5
t = np.arange(0,20,0.1)
y = odeint(model, y0, t)
plt.plot(t,y)
```

```python
y_cal = np.exp(-0.03 * t) * 5
plt.plot(t,y_cal)
```
