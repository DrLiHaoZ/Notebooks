```python
import numpy as np
```

2*2 matrix dot 1*2 vector

```python
a = np.array([-1,2])
b = np.array([[3,1],[1,2]])
print(a)
print(b)
```

```python
b.dot(a) 
```

```python
ihat = b[:,0] #i hat 
jhat = b[:,1] #j hat 
result = a[0] * ihat + a[1] * jhat
print(result)
```

```python
left_rotate = np.array([[0,-1],[1,0]])
a = np.array([1, 1])
left_rotate.dot(a)
```

to shear the vector
[1,1]
[0,1]
这个矩阵将jhat推到了(1,1)，同时ihat不变

```python
shear_jhat = np.array([[1,1],[0,1]])
a = np.array([1,1])
shear_jhat.dot(a)
```

接下来我们来看shear_yhat 与 left_rotate这两个变换顺序是否有关系

```python
a = np.array([1,1])
print(shear_jhat.dot(left_rotate.dot(a)))
print(left_rotate.dot(shear_jhat.dot(a)))
# 可以看出，顺序不同，结果也不同   
```
