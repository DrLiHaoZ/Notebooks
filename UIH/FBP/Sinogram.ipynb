```python
import tomopy
import pydicom
import matplotlib.pyplot as plt
import numpy as np

# 加载DICOM数据
ds1 = pydicom.dcmread("I:/uCT760_RunOff_WJJ_Varicosity_2_1171364_170355/1.5 x 1.2_503/00000300.dcm")
ds2 = pydicom.dcmread("I:/uCT760_RunOff_WJJ_Varicosity_2_1171364_170355/1.5 x 1.2_503/00000301.dcm")
data1 = ds1.pixel_array
data2 = ds2.pixel_array
```

```python
data = np.array([data1])
data.shape
```

```python
angles = np.linspace(0,np.pi/2,360)
projection = np.array([])
for i in range(len(angles)):
    proj = tomopy.project(data,theta=angles[i])[0,0,:]
    if i == 0:
        projection = np.hstack((projection,proj))
    else:
        projection = np.vstack((projection,proj))
```

```python
plt.imshow(projection,cmap='gray')
```
