```python
from skimage.draw import disk
import numpy as np
import matplotlib.pyplot as plt

# 创建一个空白图像
image = np.zeros((200, 200))

# 定义圆心和半径
circle_center = (100, 100)
circle_radius = 50

# 在图像中绘制圆形
rr, cc = disk(circle_center, circle_radius, shape=image.shape)
image[rr, cc] = 1.0

# 显示图像
plt.imshow(image, cmap='gray')
plt.show()
```
